# Vectorized Batch Gradient Descent

```python
def vectorizered_GD(in_x, in_y, lr=0.0005,max_iter=300):    
    ones = np.ones((in_x.shape[0],1))
    in_x = np.hstack((ones,in_x.reshape(-1,1)))
    in_y = in_y.reshape(-1,1)
    it = 0
    th = np.zeros((in_x.shape[1],1))
    h = in_x @ th
    m = in_x.shape[0]
    z = (h - in_y)**2
    cost = 1/(2*m) * z.sum()
    g = (1/m) * (in_x.T @ (h - in_y))
    all_cost = [cost]
    th0s = [th[0]]
    th1s = [th[1]]
    th = th - lr * g  
    while it < max_iter and np.linalg.norm(g) > 0.01:
#         print(cost)        
        th0s.append(th[0])
        th1s.append(th[1])
        it += 1
        h = in_x @ th
        z = (h - in_y)**2
        cost = 1/(2*m) * z.sum()
        g = (1/m) * (in_x.T @ (h - in_y))
        th = th - lr * g
        all_cost.append(cost)
        if abs(all_cost[it] - all_cost[it-1]) < 0.001:
            break
    return th, h, all_cost, th0s,th1s
t,h,costs,t0s,t1s = vectorizered_GD(x,y)
```

```python
plotting(h,t0s,t1s,x,y,costs)
```

# Mini-Batch Gradient Descent

```python
def hypothesis(t0,t1,inx):
    return t0 + t1 * inx

def compute_cost(h,iny):
    m = len(h)
    return sum([(h[i] - iny[i])**2 for i in range(m)]) / (2 * m)
```

```python
def mini_gd(in_x, in_y, batch_size, lr=0.0005, epochs=300):    
    k = len(x)//batch_size  
    theta0 = 0
    theta1 = 0
    cost = []
    th0s = []
    th1s = []
    for j in range(epochs):
        for i in range(k):
            if (i+1)*batch_size >len(x):
                mini_x = in_x[batch_size*i:]
                mini_y = in_y[batch_size*i:]
            else:    
                mini_x = in_x[batch_size*i:batch_size*(i+1)]
                mini_y = in_y[batch_size*i:batch_size*(i+1)]
            m = len(mini_x)

            h = [hypothesis(theta0,theta1,mini_x[j]) for j in range(m)]
            dt0 = sum([(h[i] - mini_y[i]) for i in range(m)]) / m
            dt1 = sum([(h[i] - mini_y[i]) * mini_x[i] for i in range(m)]) / m

            
            theta0 = theta0 - lr * dt0
            theta1 = theta1 - lr * dt1
            cost.append(compute_cost(h,mini_y))
            th0s.append(theta0)
            th1s.append(theta1)
        if np.linalg.norm([dt0,dt1]) <= 0.001:
            break
        if abs(cost[j] - cost[j-1]) < 0.01:
            break    
        
    return [theta0,theta1],cost, th0s,th1s


t,costs,t0s,t1s = mini_gd(x,y,batch_size=5)
h = t[0] + t[1] * x
```

```python
plotting(h,t0s,t1s,x,y,costs)
```
