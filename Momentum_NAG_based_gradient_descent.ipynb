```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(0,20)
y = -2*x +1
```

```python
def plotting(h,t0s,t1s,x,y,j,steps = 10):
    plt.plot(j)
    plt.show()
    
    plt.scatter(x,y)
    for i in range(0,len(costs),steps):
        y_hat = t0s[i] + t1s[i]*x    
        plt.plot(x,y_hat)
    plt.show()
    plt.scatter(x,y)
    plt.plot(x,h,c='r')
```

```python
def hypothesis(t0,t1,inx):
    return t0 + t1 * inx

def compute_cost(h,iny):
    m = len(h)
    return sum([(h[i] - iny[i])**2 for i in range(m)]) / (2 * m)
```

# Momentum-Based Gradent Descent

```python
def momentum_gradient_descent(x,y,gamma,lr = 0.0005,max_itr = 300):
    m = len(x)
    theta0= 0
    theta1= 0
    v0,v1 = 0,0
    cost = []
    th0s = []
    th1s = []
    for i in range(max_itr):
        # hypothesis
        h = [hypothesis(theta0,theta1,x[i]) for i in range(m)]
        
        # gradient
        dt0 = sum([(h[i] - y[i]) for i in range(m)]) / m
        dt1 = sum([(h[i] - y[i]) * x[i] for i in range(m)]) / m
        
        # momentum
        v0 = gamma * v0 + lr * dt0
        v1 = gamma * v1 + lr * dt1
        cost.append(compute_cost(h,y))
        
        #update thetas
        theta0 = theta0 - v0
        theta1 = theta1 - v1
        th0s.append(theta0)
        th1s.append(theta1)
#         print(f't0= {theta0_old} , t1 = {theta1_old} , h = {hypothesis(theta0_old, theta1_old,x[0])} , y = {y[0]}')
        # stops at convergence
        if np.linalg.norm([dt0,dt1]) <= 0.001:
            break
    return [theta0,theta1],h,cost, th0s, th1s
        
            
t,h,costs,t0s,t1s = momentum_gradient_descent(x,y,gamma = 0.9)
```

```python
plotting(h,t0s,t1s,x,y,costs)
```

# NAG Gradient Descent

```python
def nag_gradient_descent(x,y,gamma,lr = 0.0005,max_itr = 300):
    m = len(x)
    theta0 = 0
    theta1 = 0
    theta0 = 0
    theta1 = 0
    v0,v1 = 0,0
    cost = []
    th0s = []
    th1s = []
    for i in range(max_itr):
        #theta temp
        tt0 = theta0 - gamma * v0
        tt1 = theta1 - gamma * v1
        
        # hypothesis
        h = [hypothesis(theta0,theta1,x[i]) for i in range(m)]
        ht = [hypothesis(tt0,tt1,x[i]) for i in range(m)]
                
        # gradient
        dt0 = sum([(ht[i] - y[i]) for i in range(m)]) / m
        dt1 = sum([(ht[i] - y[i]) * x[i] for i in range(m)]) / m
        
        # look ahead step
        theta0 = tt0 - lr * dt0
        theta1 = tt1 - lr * dt1
        
        # momentum
        v0 = gamma * v0 + lr * dt0
        v1 = gamma * v1 + lr * dt1
        cost.append(compute_cost(h,y))
        
        #update thetas
        theta0 = theta0 - v0
        theta1 = theta1 - v1
        th0s.append(theta0)
        th1s.append(theta1)
#         print(f't0= {theta0} , t1 = {theta1} , h = {hypothesis(theta0, theta1,x[0])} , y = {y[0]}')
        # stops at convergence
        if np.linalg.norm([dt0,dt1]) <= 0.001:
            break
    return [theta0,theta1],h,cost, th0s, th1s
            
t,h,costs,t0s,t1s = nag_gradient_descent(x,y,gamma = 0.9)
```

```python
plotting(h,t0s,t1s,x,y,costs)
```
